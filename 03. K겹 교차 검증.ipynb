```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder

from keras.models import Sequential
from keras.layers import Dense
```

### k겹 교차 검증(k-fold cross validation)
- 데이터셋을 여러 개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 모두 합해서 학습셋으로 사용하는 방법 
- 데이터의 100%를 테스트셋으로 사용할 수 있다.

```python
# 초음파 광물 데이터 불러오기
df = pd.read_csv('data/sonar.csv', header=None)
df.head()
```

```python
# seed값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# feature, target 구성
x_data = df.drop(60, axis=1)
y_data = df[60]

x_data.shape, y_data.shape
```

```python
# 타겟 라벨 인코딩
e = LabelEncoder()
y_data = e.fit_transform(y_data)
y_data
```

```python
# k겹 교차검증(10)
# StratifiedKFold: 타켓의 비율에 맞춰서 파일을 쪼갠다
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
# 정확도를 담을 리스트
accuracy = []

# 10번 반복
# 쪼갠 train, test데이터의 인덱스 번호 반환
for train, test in skf.split(x_data, y_data):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(loss='binary_crossentropy',
                optimizer='adam',
                metrics=['accuracy'])
    
    model.fit(x_data.iloc[train], y_data[train], epochs=100, batch_size=10)
 
    acc = model.evaluate(x_data.iloc[test], y_data[test])[1]
    accuracy.append(acc)
```

```python
# 정확도
accuracy
```

```python
# 정확도 평균
np.mean(accuracy)
```
