### 유방암 데이터

```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.datasets import load_breast_cancer

from keras.models import Sequential
from keras.layers import Dense
```

```python
# 데이터 불러오기
bcancer = load_breast_cancer()
```

```python
# 데이터 정보
print(bcancer.DESCR)
```

```python
bcancer.target_names
```

```python
# 데이터 프레임 생성
bcancer_df = pd.DataFrame(bcancer.data, columns=bcancer.feature_names)
bcancer_df['target'] = bcancer.target
bcancer_df
```

```python
# 입출력 데이터
x_data = bcancer_df.drop('target', axis=1)
y_data = bcancer_df['target']

x_data.shape, y_data.shape
```

```python
# 실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(0)
tf.random.set_seed(0)
```

```python
# 딥러닝 구조 결정
model = Sequential()
# 입력층
# 30: 층 내부의 노드의 개수
# input_dim: feature 개수
# activation: 다음 층으로 어떻게 값을 넘길지 결정하는 부분
model.add(Dense(30, input_dim=30, activation='relu'))
# 출력층
# 타겟의 분류 개수(여기서는 0 또는 1이기 때문에 1)
# 다중분류면 'softmax' 사용한다
model.add(Dense(1, activation='sigmoid'))
```

```python
# 딥러닝 실행
# loss: 한 번 신경망이 실행될 때마다 오차 값을 추적하는 함수
# optimizer: 오차를 어떻게 줄여 나갈지 정하는 함수
# metrics: 모델 수행 결과를 나타내게끔 설정하는부분
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
# epochs: 반복 횟수(학습 프로세스가 모든 샘플에 대해 한 번 실행되는 것을 1epoch) 
# batch_size: 샘플을 한 번에 몇 개씩 처리할지 정하는 부분. 너무 크면 학습 속도가 느려지고, 너무 작으면 각 실행 값의 편차가 생겨서 전체 결과값이 불안정해질 수 있다.
# 컴퓨터 메모리가 감당할 만큼의 batch_size를 찾아 설정해 주는 것이 좋다
history = model.fit(x_data, y_data, epochs=100, batch_size=10)
```

```python
# 결과 출력
model.evaluate(x_data, y_data)
```
